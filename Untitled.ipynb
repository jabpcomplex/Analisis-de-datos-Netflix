# 2. CreaciÃ³n de Variables

feature extraction

```python
separacion_lista['tipo']=np.where(separacion_lista.num_partes<2,'peliculas','serie')
separacion_lista.sample(20)
```

```python
#Dado que tenemos las mismas filas y columnas solo concatenamos
df2 = pd.concat([df,separacion_lista['tipo']],axis=1)
df2.sample(10)
```

```python
separacion_cols = df.Title.str.split(pat=":",expand=True)
separacion_cols
```

```python
#para poner nombres a columnas 
separacion_cols.columns=["nivel1","nivel2","nivel3","nivel4"]
separacion_cols
```

```python
#Volvemos a concatenar
df2 = pd.concat([df2,separacion_cols],axis=1)
df2
```

# 3. Variables derivadas de la fecha

```python
df2['fecha']=pd.to_datetime(df2.Date)
#cambio la columna date de tipo object a columna(fecha) de tipo datetime64
df2
```

```python
df2.drop(columns='Date',inplace=True)
df2
```

Creamos funcion para **extraer** los componentes (**datos**)

```python
#cambiamos el indice por la fecha
df2.set_index('fecha', inplace=True)
```

```python
def atributos_fechas(data):
    data['aÃ±o'] = data.index.year  #separa por aÃ±o mes dia y agrega en df2
    data['mes'] = data.index.month_name()
    data['dia_mes'] = data.index.day
    data['dia_semana'] = data.index.day_name()
    return(data)
```

```python
atributos_fechas(df2)
#podriamos conocer que dias consumimos mas la informaciÃ³n
```

Â¿Hace cuanto tiempo hemos contratado netflix?

```python
hoy = pd.Timestamp(date.today()) #1 de abril
primer_dia = df2.index.min()
tiempo = hoy - primer_dia  #obtenemos los dias que hemos usado netflix
print(f'Elen ha usado netflix {tiempo.days} dias.\n')
print(f'El primer dÃ­a que lo uso fue {primer_dia}')
```

Â¿CuÃ¡nto tiempo de su vida le ha dedicado a Netflix

```python
#media de series y media de pelicuas

media_min_serie = 45 #minutos
media_min_pelis = 100 #minutos
consumo = df2.loc[df2.aÃ±o>2021].groupby('tipo').Title.count()
consumo
```

```python
minutos_pelis_aÃ±o = consumo['peliculas']*media_min_pelis/3
minutos_pelis_aÃ±o
```

```python
minutos_serie_aÃ±o = consumo['serie']*media_min_serie/3
minutos_serie_aÃ±o
```

```python
#convetimos minutos a dias
dias_pelis_aÃ±o = minutos_pelis_aÃ±o/60/24
dias_pelis_aÃ±o
```

```python
#convetimos minutos a dias
dias_series_aÃ±o = minutos_serie_aÃ±o/60/24
dias_series_aÃ±o
```

```python
print(f'Elen dedica al aÃ±o {round(dias_series_aÃ±o)} dias de su vida a ver series y {round(dias_pelis_aÃ±o)} dias en pelicualas' )
```
